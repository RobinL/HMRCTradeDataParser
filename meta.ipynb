```python
def convert(my_str):
    if type(my_str) == str:
        return my_str.decode('Windows-1252').encode('utf-8')
    else:
        return my_str


for row in pd.DataFrame(df.dtypes).reset_index().iterrows():
    row_name = row[1]["index"]
    row_type = row[1][0]
    
    if row_type.name == "object":
        df[row_name] = df[row_name].apply(convert)
```

```python
import re
def fix(my_str):
    my_str = my_str.replace("||","\n")
    my_str = re.sub("\n {0,10}\n", "\n", my_str)
    my_str = re.sub("\n {0,10}", "\n", my_str)
    my_str = re.sub(" {1,10}", " ", my_str)
    return my_str

df["description_fixed"] = df["Description"].apply(fix)
```
